```python
# Bubble Sort
n=5
r=np.random.random(n)*10
print("Original Array=",r)
for i in range(n):
    for j in range(n):
        if r[i]>r[j] and i<j:
            t=r[i]
            r[i]=r[j]
            r[j]=t
print("Sorted Array=",r)                            
```

```python
# Selection Sort
n=5
a=np.random.random(n)*10
print("Original Array=",a)
for i in range(n):    
    j_index = i
    for j in range(n):
        if a[j_index] > a[j] and i<j:          
            j_index = j            
    t=a[i]
    a[i]=a[j_index]
    a[j_index]=t    
print("Sorted Array=",a)             
```

```python
# Binary Search
b=np.array([4,3,7,5,73,34,45])
n=b.shape[0]
search=34

b=np.sort(b)
print("Original Array=",b)

mid=int(n/2)
stop = 0
while stop==0:    
    if b[mid]==search:
        print("Element present in position:",mid)
        stop=1
    else:
        if search<=b[mid]:      
            mid = int(0+mid/2)
        else:                   
            mid = int((mid+n)/2)
```
